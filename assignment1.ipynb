```python
# Plotting candlestick pattern for one stock
import matplotlib.pyplot as plt
xdata1 = data1[(data1['Date'] >= '2020-01-01') & (data1['Date'] <= '2023-01-31')]

plt.plot(xdata1['Date'], xdata1['High'], color='green', label='High')
plt.plot(xdata1['Date'], xdata1['Low'], color='red', label='Low')
plt.title('HINDUNILEVER High and Low Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
```

```python
# Plotting line graph for closing prices of two stocks
xdata2 = data2[(data2['Date'] >= '2020-01-01') & (data2['Date'] <= '2023-01-31')]
plt.figure()
plt.plot(xdata1['Date'], xdata1['Close'], label='HINDUNILEVER')
plt.plot(xdata2['Date'], xdata2['Close'], label='RELIANCE')
plt.title('HINDUNILEVER and RELIANCE Closing Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
```

HINDUNILEVER:
According to the first graph the bullish period of HINDUNILEVER is around the end of 2021 as both high and low prices of the stock are at its maximum and the bearish period is around mid of 2020 as the high and low prices are minimum at that point of time.Maximum high price of hindunilever is greater than 2800 whereas the minimum low price is less than 1800.
Graph 2 conveys the idea that the lowest closing price of HINDUNILEVER is in the mid of 2020 which is the bearish period whereas the maximum closing price of the stock is at the end of 2021 which states the bullish period.
The maximum closing price of hindunilever is greater than 2750 and the minimum closing price is between 1750 and 2000.
RELIANCE:
The maximum closing price of Reliance can be seen in the mid of 2022 so it is the bullish period for Reliance and minimum closing price can be seen in mid of 2020 which is the bearish period.The lowest closing price is even less tha 1000 and the highest closing price is greater than 2750.
